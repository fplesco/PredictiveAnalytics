### Import all the necessary libraries

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets as ds
```

### Import the iris dataset from sklearn package directly for use

```python
dta = ds.load_iris()

x = dta.data[:, :2]
y = dta.target
x_min, x_max = x[:, 0].min() - 0.5, x[:, 0].max() + 0.5
y_min, y_max = x[:, 1].min() - 0.5, x[:, 1].max() + 0.5

plt.figure(10, figsize=(10, 16))
plt.clf()
```

### Plot the training datapoints

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="gray")
plt.xlabel("Length")
plt.ylabel("Width")

plt.title("Plot the training datapoints")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

plt.show()
```

### Plotting the  PCA dimensions

```python
fig = plt.figure(10, figsize=(10, 16))
ax = Axes3D(fig, elev=-160, azim=110)
x_reduced = PCA(n_components=3).fit_transform(dta.data)

ax.scatter(
    x_reduced[:, 0],
    x_reduced[:, 1],
    x_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="gray",
    s=50,
)

# PCA - The First 3 Directions
ax.set_title("PCA - The First 3 Directions")

# set ticklabels for first eigenvector
ax.set_xlabel("1st Eigenvector")
ax.w_xaxis.set_ticklabels([])

# set ticklabels for second eigenvector
ax.set_ylabel("2nd Eigenvector")
ax.w_yaxis.set_ticklabels([])

# set ticklabels for third eigenvector
ax.set_zlabel("3rd Eigenvector")
ax.w_zaxis.set_ticklabels([])

# lastly, show the 3d plot
plt.show()
```
